```python
#Importar librerías
import sqlite3
import pandas as pd
import time
```

```python
# Conexión a la base de datos
conn = sqlite3.connect("libreria.db")
```

```python
# Función para ejecutar consultas y mostrar resultados bonitos
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
# Mostrar todas las tablas
run_query("SELECT name FROM sqlite_master WHERE type='table';")
```

```python
# Mostrar los primeros 10 libros
run_query("SELECT * FROM libros LIMIT 10;")
```

```python
#Cantidad de autores por libro ordenada de mayor a menor
run_query("""
SELECT libros.titulo, COUNT(libro_autor.autor_id) AS total_autores
FROM libros
LEFT JOIN libro_autor ON libros.id = libro_autor.libro_id
GROUP BY libros.titulo
ORDER BY total_autores DESC;
""")
```

```python
#Libros con más de un autor
run_query("""
SELECT libros.titulo, COUNT(libro_autor.autor_id) AS cantidad_autores
FROM libros
JOIN libro_autor ON libros.id = libro_autor.libro_id
GROUP BY libros.titulo
HAVING cantidad_autores > 1;
""")
```

```python
# Mostrar los primeros 10 libros que tengan 5 estrellas 
run_query("""
SELECT * FROM libros
WHERE estrellas='Five';
""")
```

```python
# los primeros 5 libros con 2 estrellas // CONSULTA SIN INDICE

run_query("""SELECT *
FROM libros
WHERE estrellas = 'Two'
LIMIT 5;
""")
```

```python
start = time.time()

for _ in range(100):  # repetimos 100 veces para promediar
    run_query("SELECT * FROM libros WHERE estrellas = 'Two' LIMIT 5;")

end = time.time()
print("Tiempo sin índice:", end - start, "segundos")
```
